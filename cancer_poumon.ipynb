```python
# On va regarder les valeurs aberrantes dans les colonnes numériques

# Redéfinir numerical_cols ici
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# On va utiliser la méthode de l'écart interquartile (IQR) pour détecter les valeurs aberrantes
# L'IQR est la différence entre le troisième quartile (Q3) et le premier quartile (Q1)
for col in numerical_cols:
    if col != 'id':  # On ne traite pas la colonne 'id'
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        print(f"Nombre de valeurs aberrantes dans la colonne {col} : {len(outliers)}")
```

## 5.  Partie text mining 

 ÉTAPE 1 : Chargement du texte

```python
with open("article.txt", 'r', encoding='utf-8') as fichier:
        texte = fichier.read()
        corpus = [texte]
print(f"Les corpus est composé de {len(corpus)} documents")
```

```python
corpus
```

Une autre information, peut être plus facile à exploiter serait de compter / approximer le nombre de mots dans chacun de ces documents :

```python
i = 1
for doc in corpus:
    liste_mots = doc.split() # Permet de séparer les mots d'une chaines de caractère en fonction d'un séparateur par défault l'espace.
        
    print(f"Le document {i} contient {len(liste_mots)} mots")
    print(f"La liste de mots du documents {i} : {liste_mots} \n")
    i+=1

```

 ÉTAPE 2 : Nettoyage du corpus

Le document est rédigé en français et il contient :
- des mots pouvant être conjugués et/ou accordés. Donc présence d'accents
- des chiffres
- des ponctuations
- des smileys

Notre objectifs est de normaliser ce document afin qu'il soit exploitable notamment pour des taches de visualisation ou de modélisation.

1.Le passage en minuscule

Python est sensible à la casse : *salut* est donc différent de *SALUT*. Si on souhaite par exemple, compter le nombre d'occurences du mot *salut* dans notre corpus et ce quelque soit sa place dans une phrase on va  devoir le passer en minuscule. 

La première étape consiste donc à passer l'intégralité du corpus en minuscule.

```python
# Passage du corpus en minuscule
corpus_2 = [] # initialisation de la liste

for doc in corpus:
    corpus_2.append(doc.lower()) 
    
corpus_2
```

2.Compréhension de liste et astuce

En Python, on peut utiliser les compréhensions de listes pour travailler plus facilement avec des listes. Cela permet de faire des opérations rapidement, sans devoir créer la liste à l’avance.

```python
# On va s'assurer que chaque document est bien de type str
i = 1

for doc in corpus:
    if type(doc) == str:
        print(f"Le type du document {i} est {type(doc)}, c'est ok!" )
    else :
        print(f"Le type du document {i} est {type(doc)}, il faut le convertir!" )
        
    i+=1
```

```python
# On va aussi transformer avec une compréhension de liste chaque document en chaines de caractères
corpus =  [str(doc) if type(doc)!=str else doc for doc in corpus ]

corpus # c'est ok
```

3. La suppréssion des accents

Tout comme *salut* est différent de *SALUT*, *frére* est différent de *frere*. Les machines ne disposent pas de nos capacités d'interprétation et sont donc incapables de comprendre si oui ou non il pourrait s'agir du même mot.

La fonction unidecode, qui vient du module Python du même nom, sert à remplacer les lettres accentuées par des lettres normales sans accent.


```python
# Suppréssion des accents
from unidecode import unidecode
# On va s'assurer que chaque document est bien de type str
corpus = [unidecode(doc) for doc in corpus]

corpus
```

4.   La transformation d'expression

Quand on travaille sur du texte, il faut parfois le modifier un peu pour garder les infos importantes.

Par exemple, si on a des commentaires de clients, certains laissent leur nom et prénom. Ces mots sont différents pour chaque personne, donc le modèle risque de ne pas comprendre qu’ils veulent dire "nom d'une personne".
Pour corriger ça, on peut utiliser des expressions régulières pour reconnaître les noms et les remplacer par quelque chose comme "nom_prenom".

On doit aussi souvent enlever les nombres dans le texte. Mais attention : certains chiffres, comme des dates ou des années, peuvent être utiles. Donc, il faut bien choisir quoi enlever et quoi garder.



re.sub(pattern, replacement, string) :
Cette fonction cherche toutes les occurrences qui correspondent au pattern dans la chaîne string et les remplace par replacement.  

r'[0-9]{4}' :
C’est le pattern (expression régulière). Il signifie :
* [0-9] = un chiffre entre 0 et 9
* {4} = exactement 4 fois de suite

```python
# Vous transformez toutes les suites de 4 chiffres (de 0 à 9) en "annee"
import re  # ✅ Import du module pour les expressions régulières
corpus = [re.sub(r'[0-9]{4}', 'annee', doc) for doc in corpus]

corpus
```

5. Suppression des numériques et des caractères spéciaux


Ce n’est pas toujours une bonne idée d'enlever les caractéres spéciaux d'un texte.
Par exemple, si notre texte porte sur des tweets ou des messages de réseaux sociaux, les émoticônes (smileys) peuvent être très utiles pour comprendre l’émotion. Dans ce cas-là, il vaut mieux les garder.



```python
# "[^a-z]+" ==> Une séquence d’un ou plusieurs caractères qui ne sont pas des lettres minuscules de a à z

# Les crochets [...] définissent une classe de caractères : Ça signifie « un caractère parmi ceux listés ».

# Le ^ placé juste après le crochet ouvrant [ veut dire négation : Ici, ça veut dire « tout caractère qui n’est pas dans la plage suivante ».

# a-z est la plage des lettres minuscules de l’alphabet (de a à z).

# Le + signifie « une ou plusieurs répétitions » du motif précédent.



# Vous remplacez par un espace tout ce qui est différents d'une lettre minuscule. 
corpus = [re.sub(r"[^a-z]+", ' ', doc) for doc in corpus]

corpus

```

6. Suppression des stop words

On va supprimer maintenant tous les stop words, c'est à dire tous les mots qui ne portent pas d'information pertinente compte tenu de notre sujet.

```python
# for doc in corpus :
# Pour chaque document (texte) dans ta liste corpus.

# doc.split() :
# Divise le document en mots (en utilisant l’espace comme séparateur).

# [word for word in doc.split() if word not in stopWords] :
# Crée une nouvelle liste de mots, en excluant ceux qui sont dans la liste stopWords.

# ' '.join(...) :
# Reconstitue la liste de mots filtrés en une seule chaîne de caractères, avec un espace entre chaque mot.

# Enfin, on remplace le corpus par la liste de documents nettoyés.


import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')  # à exécuter une seule fois

stopWords = set(stopwords.words('french'))  # liste des mots vides en français


corpus = [' '.join([word for word in doc.split() if word not in stopWords]) for doc in corpus]

corpus
```

7. Stemmatisation

La stemmatisation consiste à réduire un mot à sa racine (ou “stem”), c’est-à-dire à une forme de base, sans se soucier de sa forme grammaticale.

🔹 Exemples :

Mot original                                    Racine (stem)

"mange", "manger", "mangeons"	                mang
"études", "étudiant", "étudier"	                etud



```python
from nltk.stem.snowball import SnowballStemmer

# Initialisation du stemmer en français
stemmer = SnowballStemmer('french')

# Application de la stemmatisation à chaque document
corpus = [" ".join([stemmer.stem(word) for word in doc.split()]) for doc in corpus]

print(corpus[:1])  # aperçu
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Exemple : ton texte (ou ta version nettoyée si elle existe déjà)
texte = " ".join(corpus)  # corpus est ta liste de documents (ou un seul document sous forme de liste)

# Création du nuage de mots
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(texte)

# Affichage du nuage
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Nuage de mots - Cancer du poumon", fontsize=16)
plt.show()
```
